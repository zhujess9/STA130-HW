A Classification Decision Tree is a type of machine learning model that can solve classification problems. Its goal is to predict a categorical outcome based on input features by recursively splitting a dataset into smaller groups based on the values of the features. It uses conditions to maximize the 'purity' of the groups (e.g. minimizing misclassification). 

It's ideal to use the decision tree to address problems where the output is categorical, such as binary classification and multiclass classification. 
Binary Classification -> yes/no, pass/fail, true/false
Multiclass Classification -> Class A/Class B/Class C

Several examples of real-world applications can be
1. Medical Diagnosis
Predicting whether a tumor is malignant or benign based on patient features/traits (age, size, corresponding medical test results)
2. Customer Segmentation
Categorizing customers into groups such as "High Risk", "Moderate Risk", "Low Risk" for credit approval
3. Sentiment Analysis
Classifying product reviews or social media posts as positive, neutral, or negative.

Comparing Decision Tree vs Multiple Linear Regression
A classificatin decision tree makes predictions by:

Recursive Partitioning
The tree splits data into branches based on conditions that will optimize a criterion. Each branch corresponds to a subset of data with similar characteristics.

Leaf Nodes & Predictions
Once a branch ends in a 'leaf node', the majority class in that node is the predicted class for any new data point reaching it.

Non-Linear Boundaries
Classification Decision Trees can model complex, non-linear relationships in data

Multiple Linear Regressions make predictions by using:
A mathematical model that predicts a continuous outcome using y=Œ≤(not)+Œ≤1x1+Œ≤2x2+‚ãØ+Œ≤nxn+œµ where x1 are input features, Œ≤1 are coefficients, and œµ is the error term.

Linear Relationships:
It assumes a linear relationship between the input features and target variable

Output: 
The prediction is a continuous number which may later be thresholded to classify

1. **Accuracy** measures the proportion of true results (both true positives and true negatives) in the population.

$$\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}$$
    
2. **Sensitivity** measures the proportion of actual positives that are correctly identified.

$$\text{Sensitivity} = \frac{TP}{TP + FN}$$

3. **Specificity** measures the proportion of actual negatives that are correctly identified.

$$\text{Specificity} = \frac{TN}{TN + FP}$$

4. **Precision** measures the proportion of positive identifications that were actually correct.

$$\text{Precision} = \frac{TP}{TP + FP}$$

*Accuracy*
Scenario: A simple spam email classification system
Accuracy is most appropriate because of cost of false positives, such as marking a legitimate email as spam, and false negatives, such as missing a spam email is roughly the same and the classes are balanced. If spam cand non-spam emails are equally frequent, then focusing on the overall correct classification rate is reasonable.

*Sensitivity (True Positive Rate)*
Scenario: A medical screening test for a life-threatening disease such as cancer detection.
Sensitivity is critical in a situation where failing to accurately spot a positive case (false negative) will have severe consequences. High sensitivity ensures that most actual positives are correctly identified, even if it leads to some false positives

*Specificity (True Negative Rate)*
Scenario: A fraud detection systme for financial transactions.
Specificity is important when we need to correctly identify true negatives and avoid false positives. In this scenario, it can be flagging legitimate transactions as fraud. High specificity reduces inconeniences to users while ensuring normal activity is not unnecessarily disrupted.

*Precision (Positive Predictive Value)*
Scenario: A targeted advertising system predicting whether high-value customers are likely to purchase a product.
Precision will be important when we care about the correctedness of positive predictions and minimize falsse positives. Incorrectly targeting non-buyers will waste advertisement resources and can harm the brand's reputation, so, precision will ensure that predicted positives are most likely true positives.

Predictions: For a new book, the tree will follow the splits based on its List Price and arrive at a leaf node where the class with majority determines the prediction

> The way a **classification decision tree** is fit is that at each step in the construction process of adding a new **decision node splitting rule** to the current **tree structure**, all possible **decision rules** for all possible **predictor variables** are considered, and the combination that improves the **prediction** the most (as measured by the criterion of either "Gini impurity" or "Shannon entropy") and in accordance with the rules of the decision tree (such as the `max_depth` argument) is added to the **classification decision tree**.  Thus overall "criterion" noted above improves with each new **decision node splitting rule**, so the improvement can thus be tracked and the improvement contributions attributed to the **feature** upon which the **decision node splitting rule** is based.  This means the relative contribution of each **predictor variable** to the overall explanatory power of the model can be calculated, and this is what the `.feature_importances_` attribute does. 
>
> Compared to the simplicity of understanding how different **covariates** contribute towards the final **predicted values** of **multiple linear regression models** (by just reading off the equation to see how predictions work), the the complexity of how all the different **features** interact and combine to together to create the final **predictions** from **classification decision trees** can be staggering. But the so-called **feature importance** heuristics allows us to judge how relatively important the overall contributions from different features are in the final decision tree predictions. Now we just need to be sure we're not **overfitting** our **classification decision trees** since they can be so **complex**.
